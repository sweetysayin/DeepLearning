```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics import f1_score,classification_report
from sklearn import datasets
data=datasets.fetch_20newsgroups()
print(data.DESCR)
categories=['rec.sport.hockey','rec.sport.baseball','rec.autos']
data=datasets.fetch_20newsgroups(subset='train',categories=categories,remove=('headers','footers','quotes'))
```

```python
newsgroups_test=datasets.fetch_20newsgroups(subset='test',categories=categories,remove=('headers','footers','quotes'))
```

```python
print(type(data))
print(type(data.data))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import Perceptron
vectorizer=TfidfVectorizer()
x_train=vectorizer.fit_transform(data.data)
x_test=vectorizer.transform(newsgroups_test.data)
```

```python
#print(x_train)
print(x_train.ndim)
print(x_train.shape)
#print(x_test)
print(x_test.ndim)
print(x_test.shape)

```

```python
classifier=Perceptron(n_iter_no_change=100,eta0=0.1)
classifier.fit(x_train,data.target)
```

```python
predictions=classifier.predict(x_test)
predictions
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(newsgroups_test.target,predictions))
```
